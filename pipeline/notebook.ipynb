```python
# 1. ALWAYS strip trailing/leading spaces first so things like 'RABAT ' or 'CASABLANCA ' are caught
df['City'] = df['City'].str.strip()

# 2. Your complete, comprehensive mapping list
mapping = {
    # Variations of existing Moroccan Cities (Accents & Typos)
    'FEZ': 'FES',
    'FÈS': 'FES',
    'TANGER': 'TANGIER',
    'TANGER-ASSILAH': 'TANGIER',
    'MOHAMMADIA': 'MOHAMMEDIA',
    'BERRCHID': 'BERRECHID',
    'CASZBLANCA': 'CASABLANCA',
    'MARRAKESH': 'MARRAKECH',
    'MEKNÈS': 'MEKNES',
    'TÉTOUAN': 'TETOUAN',
    'KÉNITRA': 'KENITRA',
    'KHÉMISSAT': 'KHMISSAT', # Unifies KHÉMISSAT and KHEMISSET to a single clean standard
    'KHEMISSET': 'KHMISSAT',
    'SALÉ': 'SALE',
    'MÁLAGA': 'MALAGA',
    
    # Ben Guerir variations
    'BENGUERIR': 'BEN GUERIR',
    'BENGUERIR (THE GREEN CITY OF BENGUERIR)': 'BEN GUERIR',
    'BEN_GUERIR': 'BEN GUERIR',
    
    # Beni Mellal variations
    'BÉNI MELLAL': 'BENI MELLAL',
    'BENI_MELLAL': 'BENI MELLAL',
    
    # Skhirate-Temara standard
    'SKHIRATE-TÉMARA': 'TEMARA',
    
    # Multi-City or Contextual Cleanups (Maps them to a primary local city)
    'CASABLANCA, MOROCCO': 'CASABLANCA',
    'ABU DHABI + RABAT BASED SARL': 'RABAT',
    'CASABLANCA - MARRAKECH AND FES': 'CASABLANCA',
    'CASABLANCA/DUBAI/LONDON': 'CASABLANCA'
}

# 3. Apply the mapping
df['City'] = df['City'].replace(mapping)
```

```python
import unicodedata

# 1. Clean up structural noise first (spaces, casing, punctuation)
def pre_clean_sector(text):
    if not isinstance(text, str):
        return text
    
    # Strip whitespace and convert to upper case
    text = " ".join(text.split()).upper()
    
    # Replace underscores with spaces (handles CLEAN_TECH, INSUR_TECH, etc.)
    text = text.replace('_', ' ')
    
    # Strip French accents programmatically (handles 'sécurité des systémes' -> 'SECURITE DES SYSTEMES')
    text = ''.join(c for c in unicodedata.normalize('NFD', text) if unicodedata.category(c) != 'Mn')
    
    # Remove apostrophes or weird spacing around them (e.g., dinformation -> D INFORMATION)
    text = text.replace("DINF", "D INF") 
    
    return text

# Apply pre-cleaning
df['Main sector'] = df['Main sector'].apply(pre_clean_sector)

# 2. Map the unified variations to a clean, standardized industry standard
sector_mapping = {
    # French to English unification
    'SPORT ELECTRONIQUE': 'SPORTS TECH',
    'SECURITE DES SYSTEMES D INFORMATION': 'CYBERSECURITY',
    
    # Snake_case/Abbreviation cleanups (now capitalized and space-separated)
    'CLEAN TECH': 'CLEANTECH',
    'CONSTRUTECH': 'PROPTECH',       # Construction tech often rolls into PropTech, or keep as CONSTRUTECH
    'TREE D TECHNOLOGY': '3D TECH',  # Fixes phonetic 'Tree D' typo
    'LEGAL TECH': 'LEGALTECH',
    'INSUR TECH': 'INSURTECH',
    'MED TECH': 'HEALTHTECH',         # Standardizes Medical Tech into HealthTech
    
    # Long parenthetical entries cleanup
    'MARTECH (MARKETING / ADVERTISING /DIGITAL & SOCIAL MEDIA)': 'MARTECH',
    'IMMERSIVE TECH (AR, VR, MR, GAMING)': 'IMMERSIVE TECH',
    'LEADS GENERATION OPTIMIZATION': 'MARTECH' # Leads optimization typically falls under MarTech
}

# Apply final mapping
df['Main sector'] = df['Main sector'].replace(sector_mapping)
```
